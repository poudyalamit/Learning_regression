```python
from sklearn.datasets import make_circles
import matplotlib.pyplot as plt

X,y = make_circles(n_samples=200,shuffle=True,random_state=42,noise=0.1)

plt.scatter(X[:,0],X[:,1],c=y)
plt.show()
```

```python
from sklearn.datasets import make_moons
Z,y = make_moons(n_samples=200,shuffle=True,random_state=42,noise=0.1)

plt.scatter(Z[:,0],Z[:,1],c=y)
plt.show()
```

```python
from sklearn.datasets import make_classification

X,y= make_classification(n_samples=100,n_features=2,n_redundant=0,n_informative=2,n_repeated=0,n_classes =3, n_clusters_per_class=1)

plt.scatter(X[:,0],X[:,1],c=y)
plt.show()
```
